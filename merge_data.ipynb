```python
import pandas as pd
import glob, os

dfs = [pd.read_csv(fp).assign(New=os.path.basename(fp).split('.')[0]) for fp in glob.glob('./bottom_up_signals/*.csv')]
df = pd.concat(dfs, ignore_index=True)
df.to_parquet("./bottom_up_signals.parquet")
```
